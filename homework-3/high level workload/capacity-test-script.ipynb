```python
#utilities
import matplotlib.pylab as plt
import pandas as pd

# def variabili

thrput = {10: 59.49, 
          50: 101.64, 
          100: 186.58, 
          200: 363.02, 
          300: 502.29, 
          400: 674.74, 
          500: 814.68, 
          600: 964.50, 
          700: 1122.08, 
          800: 1228.99, 
          1200: 1508.08, 
          2000: 1580.69
         }

resp = {10: 0.0010, 
        50: 0.0010, 
        100: 0.0012, 
        200: 0.0012, 
        300: 0.0012, 
        400: 0.0013, 
        500: 0.0014, 
        600: 0.0016, 
        700: 0.0017, 
        800: 0.0018, 
        1200: 0.0055, 
        2000: 0.0071
       }
```

```python
thr_num = thrput.items()
resp_num = resp.items()
thr_num = sorted(thr_num)
resp_num = sorted(resp_num)
x1,y1 = zip(*thr_num)
x2, y2 = zip(*resp_num)

plt.plot(x1,y1)
plt.xlabel("CTT")
plt.ylabel("throughput")
plt.show()

plt.plot(x2,y2)
plt.xlabel("CTT")
plt.ylabel("response time")
plt.show()

power = {}
for i in thrput.keys():
    power[i] = thrput[i] / resp[i]


pwr_num = power.items()
pwr_num = sorted(pwr_num)
x3,y3 = zip(*pwr_num)
plt.plot(x3,y3)
plt.xlabel("CTT")
plt.ylabel("power")
plt.show()
```
